```python
import numpy as np
from random import choice
import pandas as pd
import random
import argparse
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # 3D drawing in space
from pylab import *
mpl.rcParams['font.sans-serif'] = ['SimHei']
matplotlib.rcParams.update({'font.size': 22})
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
plt.rcParams.update({'font.size': 15})
colors = ['red', 'green', 'blue']
```

### 转移时间对比

```python
simulation_times_list = []
HPs = [2,3,5,10,20]
for HP in HPs:
    simulation_times = np.load('log/task3_simulation_times_hp%d.npy'%HP)
    simulation_times_list.append(simulation_times)
```

```python
plt.figure(figsize=(15,5), dpi = 144)
plt.title('Analysis of competition times')
for idx, simulation_times in enumerate(simulation_times_list):
    plt.plot(list(range(len(simulation_times))), simulation_times, color=colors[idx%3], label='Program simulation with HP=%d'%HPs[idx], linewidth = idx+1)
plt.legend()
plt.xlabel('Index of non recurrent state')
plt.ylabel('Number of competition')
plt.savefig('figure/Task3-Number of competition 2D view comparison.svg')
plt.savefig('figure/Task3-Number of competition 2D view comparison.png')
plt.show() 
```

## 转移概率对比

```python
simulation_probabilitys_list = []
HPs = [2,3,5,10,20]
for HP in HPs:
    simulation_probabilitys = np.load('log/task3_simulation_probabilitys_hp%d.npy'%HP)
    simulation_probabilitys_list.append(simulation_probabilitys)
```

```python
plt.figure(figsize=(15,5), dpi = 144)
plt.title('Analysis of final coin distribution (average coin possibility distribution from all start state)')
for idx, HP in enumerate(HPs):
    simulation_probabilitys = simulation_probabilitys_list[idx]
    plt.plot(list(range(simulation_probabilitys.shape[1])), np.mean(simulation_probabilitys, axis=0), color=colors[idx%3], label='Program simulation with HP=%d'%HPs[idx], linewidth = idx+1)
plt.legend()
plt.xlabel('Index of valid absorbing states')
plt.ylabel('Average probability of coin distribution')
plt.savefig('figure/Task3-Average probability of coin distribution 2D view comparison.svg')
plt.savefig('figure/Task3-Average probability of coin distribution 2D view comparison.png')
plt.show() 
```
